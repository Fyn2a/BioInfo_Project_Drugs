```python
BioActivity = new_client.activity
residues2 = BioActivity.filter(target_chembl_id=target_selection).filter(standard_type='IC50')
df2 = pd.DataFrame.from_dict(residues2)
```

## What is IC50?
An IC50 measurement tells us the concentration at which a drug is able to inhibit a particular biological process by 50%. For example, if compound A can inhibit 50% of the binding of a ligand to a particular receptor at a concentration of 5 nM, it would have an IC50 value of 5 nM.

source: https://www.collaborativedrug.com/cdd-blog/why-using-pic50-instead-of-ic50-will-change-your-life

```python
df2.to_csv("Bioactivity.csv", index = False)
```

```python
df3 = df2[df2['standard_value'].notna()]
df3
```

```python
bioactivity_class = []
for i in df3.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")

selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df4 = df3[selection]
df4
    
```

```python
bioactivity_class = pd.Series(bioactivity_class, name='bioactivity_class')
df5 = pd.concat([df4, bioactivity_class], axis=1)
df5
```

```python
df_cleaned = df5.dropna(subset=['molecule_chembl_id', 'standard_value','bioactivity_class'])
df_cleaned = df_cleaned.reset_index(drop=True)
df_cleaned
```

```python
df_cleaned.to_csv('Bioactivity_preprocessed.csv', index=False)
```
