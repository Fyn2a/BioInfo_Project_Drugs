## Data Generation.

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
target = new_client.target
taget_importing = target.search('dengue')
list_of_targets = pd.DataFrame.from_dict(taget_importing)
list_of_targets
```

```python
target_selection = list_of_targets.target_chembl_id[5]
target_selection
```

### Bioactivity data

```python
BioActivity = new_client.activity
residues = BioActivity.filter(target_chembl_id=target_selection)
df = pd.DataFrame.from_dict(residues)
df.head(1)
```

```python
print(df.columns.tolist())
```

```python
print(df['standard_type'].unique())
print(df['standard_type'].value_counts()['IC50'])
```

```python
residues2 = BioActivity.filter(target_chembl_id=target_selection).filter(standard_type='IC50')
df2 = pd.DataFrame.from_dict(residues2)
df2.head(3)
```

```python
df2.to_csv("Bioactivity.csv", index = False)
```
